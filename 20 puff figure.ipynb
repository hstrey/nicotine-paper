```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
```

```python
puff20=pd.read_csv("20Puff.csv",index_col=0)
```

```python
puff20
```

```python
puff20t = puff20.transpose()
puff20t
```

```python
plot_x = ["Baseline","4","8","12","16","20","EOS"]
num_columns = ['Age','Weight (lb)','Height ("")']
puff20t[plot_x]=puff20t[plot_x].apply(pd.to_numeric)
puff20t[num_columns]=puff20t[num_columns].apply(pd.to_numeric)
# create new column "offset" and set value to zero if value is 1 (one could mean anything below 2)
puff20t["offset"]=puff20t["Baseline"]
puff20t.loc[puff20t["offset"]==1,"offset"]=0
puff20t[plot_x].plot(kind='box')
```

```python
# create body weight normalized nicotine concentrations
plot_x_norm = ["BaselineN","4N","8N","12N","16N","20N","EOSN"]
puff20t[plot_x_norm]=puff20t[plot_x].divide(puff20t['Weight (lb)'],axis=0)*150
puff20t[plot_x_norm].plot(kind='box')
```

```python
nic_norm_mean=np.array(puff20t[plot_x_norm].mean())
nic_norm_std=np.array(puff20t[plot_x_norm].std())
puff_norm=np.array([0,4,8,12,16,20])
p=np.polyfit(puff_norm,nic_norm_mean[:-1],1)
print(p)
linefit = p[1]+puff_norm*p[0]
```

```python
plt.errorbar(puff_norm,nic_norm_mean[:-1],yerr=nic_norm_std[:-1], marker='o')
plt.plot(puff_norm,linefit)
plt.xlim((-0.3,20.3))
plt.xlabel('Sprays')
plt.ylabel('normalized blood nicotine concentration ng/ml')
```

```python
# try something different.  First subtract the offset and then normalize
plot_x_norm_offset = ["BaselineON","4ON","8ON","12ON","16ON","20ON","EOSON"]
puff20t[plot_x_norm_offset]=puff20t[plot_x].subtract(puff20t["offset"],axis=0).divide(puff20t['Weight (lb)'],axis=0)*150
puff20t[plot_x_norm_offset].plot(kind='box')
```

```python
nic_norm_offset_mean=np.array(puff20t[plot_x_norm_offset].mean())
nic_norm_offset_std=np.array(puff20t[plot_x_norm_offset].std())
puff_norm_offset=np.array([4,8,12,16,20])
p2=np.polyfit(puff_norm_offset,nic_norm_offset_mean[1:-1],1)
print(p2)
linefit = p2[1]+puff_norm_offset*p2[0]
```

```python
plt.errorbar(puff_norm_offset,nic_norm_offset_mean[1:-1],yerr=nic_norm_offset_std[1:-1], marker='o')
plt.plot(puff_norm_offset,linefit)
plt.xlim((-0.3,20.3))
plt.xlabel('Sprays')
plt.ylabel('normalized blood nicotine concentration ng/ml')
```
