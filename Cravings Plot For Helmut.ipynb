```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
for i in range(1,5):
    
    globals()['subject{}'.format(i)] = sio.loadmat('Adlib'+str(i)+ '.mat')
    globals()['crav{}'.format(i)] = globals()['subject{}'.format(i)]['Cravings'].flatten()
    globals()['time{}'.format(i)] = globals()['subject{}'.format(i)]['Time'].flatten()
    globals()['puffs{}'.format(i)] = globals()['subject{}'.format(i)]['Puffs'].flatten()
```

```python
my_labels = {"x1" : "Self-Administered Nicotine Puff"}
plt.figure(figsize=(25,10))
plt.plot(time1,crav1)
for p in puffs1[1:]:
    plt.axvline(x=p,color='r',label=my_labels["x1"])
    my_labels["x1"] = "_nolegend_"
plt.legend(fontsize=30,loc=9, bbox_to_anchor=(0.793, 1.13))
   
plt.xlim(0,3000)
plt.ylim(1,125)
plt.xlabel('Time (seconds)',fontsize=30)
plt.ylabel('Self-Reported Cravings',fontsize=30)
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)
plt.title('Subject '+ str(1),fontsize=30)
plt.savefig('Subject1'+'.pdf')
```

```python
group = sio.loadmat('GroupPlot.mat')
crav_t = group['Time'].flatten()
crav_v = group['MeanCraving'].flatten()
puffs = group['Puffs'].flatten()

my_labels = {"x1" : "Nicotine Puff"} 

plt.figure(figsize=(25,10))
plt.plot(crav_t,crav_v)
plt.xlim(0,3000)
plt.ylim(1,125)
plt.xlabel('Time (seconds)',fontsize=30)
plt.ylabel('Mean Self-Reported Cravings',fontsize=30)
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)
plt.title('Group Plot', fontsize=30)
# Overlay Puffs
for p in puffs:
    plt.axvline(x=p,color='r',label = my_labels["x1"])
    my_labels["x1"] = "_nolegend_"
plt.legend(fontsize=30,loc=9, bbox_to_anchor=(0.893, 1.13))
plt.savefig('Group Plot'+'.pdf')
```
